```python
# import libraries

from IPython.display import HTML

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# step 1.3: import train.csv into a data frame

train_file = 'train.csv'
df = pd.read_csv(train_file)
```

```python
# number of rows and columns in the data frame

df.shape
```

```python
# check which columns have null values

df.isnull().sum()
```

```python
# step 2.1: create a bar chart showing how many missing values are in each column

df.isna().sum().plot(kind='bar')
```

```python
# step 2.2: which col has the most null values? how many cells are empty?
# Cabin; 687 cells
```

```python
# step 2.3: delete all rows where "Embarked" is empty

df.dropna(subset=['Embarked'], inplace=True)
df.shape #891 rows - 2 rows = 889 rows
```

```python
# return rows where value for Cabin column is null
df[df.Cabin.isnull()]
```

```python
# step 2.4: fill all empty cabins

df['Cabin'].fillna('**¯\\_(ツ)_/¯**', inplace=True)
df
```

```python
# step 3.1: create a new column called 'FamilyCount'

df['FamilyCount'] = df[['SibSp', 'Parch']].sum(axis = 1)
df
```

```python
# step 3.2: create a new column called 'IsReverend'

df['IsReverend'] = df['Name'].map(lambda name: 1 if "Rev." in name else 0)
df.iloc[145:155]
```

```python
# step 3.3a: create three new columns 'Embarked_C', 'Embarked_Q', 'Embarked_S'

df2 = pd.get_dummies(df, columns=['Embarked'],dtype=int)
```

```python
# step 3.3b: create two new columns 'Sex_male', 'Sex_female'

df3 = pd.get_dummies(df2, columns=['Sex'],dtype=int)
```

```python
# step 3.3c: create 'Title' column (BONUS)

df3['Title'] = df3['Name'].map(lambda name: name.split(',')[1].split(' ')[1])
df3.head()
```

```python
# total passengers, total survivors, total victims

total_passengers = df3['PassengerId'].count()
total_survivors = df3['Survived'].value_counts()[1]
total_victims = df3['Survived'].value_counts()[0]

print(f"Total passengers is {total_passengers}.")
print(f"Total survivors is {total_survivors}.")
print(f"Total victims is {total_victims}.")
```

```python
# step 4.1: overall survival rate

survival_rate = f"{total_survivors / total_passengers:.1%}"
print(f"Overall survival rate is {survival_rate}.")
```

```python
# step 4.2: which gender fared the worst

print(f"Male survival rate is {df3.groupby('Sex_female')['Survived'].mean()[0]:.1%}.")
print(f"Female survival rate is {df3.groupby('Sex_female')['Survived'].mean()[1]:.1%}.")
```

```python
# step 4.3: survival rate for each Pclass, shorter method

print(f"Class 1 survival rate is {df3.groupby('Pclass')['Survived'].mean()[1]:.1%}.")
print(f"Class 2 survival rate is {df3.groupby('Pclass')['Survived'].mean()[2]:.1%}.")
print(f"Class 3 survival rate is {df3.groupby('Pclass')['Survived'].mean()[3]:.1%}.")
```

```python
# step 4.4: how many reverends survived

df_rev = df3[df3["Name"].str.contains("Rev.")][['Survived','Name']]
print(df_rev)
print("No reverends survived.")
```

```python
# step 4.5: survival rate for cabins marked **¯\\_(ツ)_/¯**

df_cabin = df3[df3["Cabin"].str.contains("_")]
cabin_survived = df_cabin['Survived'].value_counts()[1]
cabin_survival_rate = f"{cabin_survived / df_cabin['PassengerId'].count():.1%}"
print(f"Survival rate for cabins marked **¯\\_(ツ)_/¯** is {cabin_survival_rate}.")
```

```python
# step 4.6: survival rate for passengers whose age is empty

df_age = df3[df3["Age"].isnull()]
empty_age_survived = df_age['Survived'].value_counts()[1]
age_survival_rate = f"{empty_age_survived / df_age['PassengerId'].count():.1%}"
print(f"Survival rate for passengers with null age is {age_survival_rate}.")
```

```python
# step 4.7: survival rate for each port of embarkation, shorter method

print(f"Port C survival rate is {df.groupby('Embarked')['Survived'].mean()[0]:.1%}.")
print(f"Port Q survival rate is {df.groupby('Embarked')['Survived'].mean()[1]:.1%}.")
print(f"Port S survival rate is {df.groupby('Embarked')['Survived'].mean()[2]:.1%}.")
```

```python
# step 4.8: survival rate for children under 12 in each pclass, shorter method

df_child1 = df3.drop(df3[df3['Age'] >= 12].index)
df_child2 = df_child1.drop(df3[df3['Age'].isnull()].index)

child_survived = df_child2.groupby(['Pclass','Survived'])['PassengerId'].count()
child_all = df_child2.groupby(['Pclass'])['PassengerId'].count()

print(f"Class 1 survival rate for children is {child_survived[1][1] / child_all[1]:.1%}.")
print(f"Class 2 survival rate for children is {child_survived[2][1] / child_all[2]:.1%}.")
print(f"Class 3 survival rate for children is {child_survived[3][1] / child_all[3]:.1%}.")
```

```python
# step 4.9: did captain survive

df_rev = df3[df3["Title"] == "Capt."][['Survived','Name']]
print(df_rev)
print("The captain did not survive.")
```

```python
# step 4.10: who died and had the most expensive ticket

df_dead = df3[df3['Survived'] == 0]
print("Mr. Charles Alexander Fortune and Mr. Mark Fortune had the most expensive fare of $263 amongst the deceased.")
print(df_dead[['Name','Fare']].sort_values(by='Fare', ascending=False))
```

```python
# step 4.11: does having family help or hurt your chances of survival

survive_w_fam = df3[df3['FamilyCount'] > 0]['Survived'].mean()
survive_wout_fam = df3[df3['FamilyCount'] == 0]['Survived'].mean()

print(f"Survival rate for passengers with family onboard is {survive_w_fam:.1%}.")
print(f"Survival rate for passengers with no family onboard is {survive_wout_fam:.1%}.")
```

```python
# step 5a: survival count by pclass

df.groupby('Pclass').count()['Survived'].plot(kind='bar', rot=0)
plt.xlabel('Passenger Class');
plt.ylabel('Number of Survivors');
```

```python
# step 5b: survival count by sex

df.groupby('Sex').count()['Survived'].plot(kind='bar', rot=0)
plt.xlabel('Gender');
plt.ylabel('Number of Survivors');
```

```python
# step 5c: survival count by port of embarkation

df.groupby('Embarked').count()['Survived'].plot(kind='bar', rot=0)
plt.xlabel('Port of Embarkation');
plt.ylabel('Number of Survivors');
```

```python
# step 5d: age distribution by survival status

df.boxplot(column='Age', by='Survived');
plt.xlabel('Survival Status');
plt.ylabel('Age');
```

```python
# step 5e: fare distribution by survival status

df.boxplot(column='Fare', by='Survived');
plt.xlabel('Survival Status');
plt.ylabel('Fare Paid');
```
